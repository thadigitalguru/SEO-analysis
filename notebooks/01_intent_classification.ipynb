# 01 — Intent Classification (Zero‑Shot)
Goal: Tag keywords as **informational, navigational, commercial, transactional**.

### Logic
Zero‑shot compares your text against label hypotheses using an NLI model. Great when you don't have labeled data yet.

```python
from transformers import pipeline
import pandas as pd
labels = ['informational','navigational','commercial','transactional']
clf = pipeline('zero-shot-classification', model='facebook/bart-large-mnli')
df = pd.read_csv('data/examples/keywords.csv')
df['intent'] = df['keyword'].apply(lambda kw: clf(kw, labels)['labels'][0])
df.head()
```
