Identify ORG/LOC/PER entities in SERP snippets to inform briefs & structured data.

```python
from transformers import pipeline
import pandas as pd
nlp = pipeline('ner', model='dslim/bert-base-NER', aggregation_strategy='simple')
df = pd.read_csv('data/examples/serp_samples.csv')
text = df['title'].iloc[0] + '. ' + df['snippet'].iloc[0]
nlp(text)
```
