# 02 â€” Keyword Clustering with Embeddings
Embed with `all-MiniLM-L6-v2` and group with KMeans to uncover topic clusters.

```python
from sentence_transformers import SentenceTransformer
from sklearn.cluster import KMeans
import pandas as pd
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
df = pd.read_csv('data/examples/keywords.csv')
X = model.encode(df['keyword'].tolist(), show_progress_bar=True, normalize_embeddings=True)
km = KMeans(n_clusters=5, n_init=10, random_state=42).fit(X)
df['cluster'] = km.labels_
df.head()
```
