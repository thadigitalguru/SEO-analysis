Small T5 model to suggest titles and meta descriptions.

```python
from transformers import pipeline
gen = pipeline('text2text-generation', model='google/flan-t5-small')
prompt = 'Generate an SEO title and 150-160 char meta for: best running shoes'
gen(prompt, max_new_tokens=96)
```
