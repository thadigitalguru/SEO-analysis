# 06 â€” Tiny Semantic Search
Index a small corpus and retrieve by meaning (embeddings + nearest neighbors).

```python
from sentence_transformers import SentenceTransformer
from sklearn.neighbors import NearestNeighbors
import pandas as pd, json
df = pd.read_csv('data/examples/serp_samples.csv')
texts = (df['title'] + '. ' + df['snippet']).tolist()
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
X = model.encode(texts, show_progress_bar=True, normalize_embeddings=True)
nn = NearestNeighbors(n_neighbors=3, metric='cosine').fit(X)
def search(q):
    qx = model.encode([q], normalize_embeddings=True)
    dists, idx = nn.kneighbors(qx, n_neighbors=3)
    return [texts[i] for i in idx[0]]
print(json.dumps(search('marathon shoes for beginners'), indent=2))
```
